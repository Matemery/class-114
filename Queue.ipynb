```python
numbers = [1,2,3,4,5]
queue = []

for number in numbers:
    queue.append(number)
    
new_queue= []    
    
while len(queue) >0:
    item = queue.pop(0)
    new_queue.insert(0, item)
    
print(new_queue)
```
