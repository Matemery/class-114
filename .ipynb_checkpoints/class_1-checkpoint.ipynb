# problem solving and AAA Approach 

AAA Approach (Assessment, Assembly, Action):
- *Assessment:* We want to ask meaningful questions with a purpose of clarifying incorrect assumption, edge cases we might have missed and specific criteria we do not know up front.

- *Assembly*: We want to design a solution by defining data structures, selecting or defining and trying to solve the problem as optimally as possible,

- *Action*: Code.
